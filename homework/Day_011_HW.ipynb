```python
import pandas as pd
q_df = pd.DataFrame([['male', 'teacher'], 
              ['male', 'engineer'],
              ['female', None],
              ['female', 'engineer']],columns=['Sex','Profession'])
q_df
```

```python
#缺失值填入字串'others'
new_q_df = q_df.fillna("others")
new_q_df
```

```python
#更進一步將字串做編碼。 此時用什麼方式做編碼比較適合?為什麼?
q_label_pro = pd.get_dummies(new_q_df['Profession'])
new_q_df = pd.concat([new_q_df,q_label_pro],axis=1)
new_q_df
```

```python
from sklearn.preprocessing import LabelEncoder
new_q_df['sex_label'] = LabelEncoder().fit_transform(new_q_df['Sex'].values)
new_q_df
```

```python
#1.對於Profession這種多類別特徵的欄位，可採one-hot encoding以轉dummy variables做分析
#2.對於Sex這種二分法的欄位，可採LabelEncoder轉為布林值做分析
```
