```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('C:/Users/angelahou/Documents/GitHub/boston.csv',index_col = 0)
print(data.head(5))#確認資料內容有哪些欄位

#1.畫出箱型圖，並判斷哪個欄位的中位數在 300~400 之間？
data.boxplot(figsize = (10,8))

print("")
print("驗證中位數：")
c1 = data.median()#所有欄位的中位數
c2 = c1[300 < c1] #大於300
c3 = c2[c2 < 400] #小於400
print(c3.index,"前述欄位的中位數位於300~400之間")
```

```python
#2.畫出散佈圖 x='NOX', y='DIS' ，並說明這兩欄位有什麼關係？
data.plot.scatter(x="NOX",y="DIS")
```

```python
print("由上圖可知，NOX與DIS欄位呈現反比關係，且非為線性關係")
```
